```python
def format_ocr_output(text):
    """
    General-purpose formatter for raw OCR text.
    Classifies each line into categories based on content.
    """
    lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

    result = {
        "Names / Initials": [],
        "Numbers": [],
        "Alphanumeric Codes": [],
        "Special Characters / Symbols": [],
        "Keywords / Labels": [],
        "Unclassified": []
    }

    for line in lines:
        if line.isdigit():
            result["Numbers"].append(line)
        elif line.isalpha() and len(line) <= 4:
            result["Names / Initials"].append(line)
        elif line.replace(" ", "").isalnum():
            result["Alphanumeric Codes"].append(line)
        elif any(char in line for char in "!@#$%^&*()_+={}[]|\\:;\"'<>,.?/~`"):
            result["Special Characters / Symbols"].append(line)
        elif any(keyword in line.lower() for keyword in ["date", "name", "age", "temp", "pulse", "id", "number", "address", "center", "hospital"]):
            result["Keywords / Labels"].append(line)
        else:
            result["Unclassified"].append(line)

    return result
```

```python
def format_ocr_output(text):
    lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

    result = {
        "Names / Initials": [],
        "Numbers": [],
        "Alphanumeric Codes": [],
        "Special Characters / Symbols": [],
        "Keywords / Labels": [],
        "Unclassified": []
    }

    for line in lines:
        if line.isdigit():
            result["Numbers"].append(line)
        elif line.isalpha() and len(line) <= 4:
            result["Names / Initials"].append(line)
        elif line.replace(" ", "").isalnum():
            result["Alphanumeric Codes"].append(line)
        elif any(char in line for char in "!@#$%^&*()_+={}[]|\\:;\"'<>,.?/~`"):
            result["Special Characters / Symbols"].append(line)
        elif any(keyword in line.lower() for keyword in ["date", "name", "age", "temp", "bp", "id", "number", "address", "timing", "mob", "ph", "medicine", "dosage", "duration", "follow", "advice"]):
            result["Keywords / Labels"].append(line)
        else:
            result["Unclassified"].append(line)

    return result


# ✔️ OCR Text Input as Multiline String
raw_text = """
Dr. Onkar Bhave Care Clinic
M.B.BS, M.D, M.S. | Reg. No: 270988 | Near Axis Bank, Kothrud, Pune
Mob. No: 8983390126 vomtfine 411038,
Ph: 094233 80390, Timing: 09:00 AM - 02:00 PM | Closed: Thursday

ID: 266 - DEMO PATIENT (M)
Address: PUNE
Temp (deg): 36, BP: 120/80 mmHg
Date: 27-Apr-2020, 04:37 PM

R
Medicine Name Dosage Duration
1. betnovate 30mg (Before Food) 1 Morning, 1 Night 10 Days (Tot: 20 Tab)
2. alkasol 30mg (Before Food) 1 Morning, 1 Night 40 Days (Tot: 20 cap)
3. paracetamol 30mg (After Food) 1 Morning, 1 Aft, 1 Eve, 1 Night 10 Days (Tot: 40 Tab)

Advice Given:
AVOID OILY AND SPICY FOOD
Follow Up: 12-05-2020
"""

# 🚀 Run formatter
formatted = format_ocr_output(raw_text)

# 📋 Display results
for category, items in formatted.items():
    print(f"\n🔹 {category}:")
    for item in items:
        print(f"   - {item}")
```

```python
def parse_easyocr_ocr_output(text):
    lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

    # Structure to hold interpreted values
    result = {
        "Patient ID / Number": None,
        "Center Info": None,
        "Important Codes": [],
        "Extra Info": []
    }

    for line in lines:
        if line.isdigit() and len(line) >= 5:
            # Likely to be an ID or patient number
            if not result["Patient ID / Number"]:
                result["Patient ID / Number"] = line
            else:
                result["Important Codes"].append(line)
        elif any(keyword in line.lower() for keyword in ["center", "cntr", "hospital", "clinic", "cntr &"]):
            result["Center Info"] = line
        elif any(char.isalnum() for char in line):
            result["Important Codes"].append(line)
        else:
            result["Extra Info"].append(line)

    return result


# ✅ Input example (multiline string from EasyOCR output)
ocr_output = """
Aec
894271
41-
2
n
4
JV
T~td
647
I_
D"
34
Jtteo
Lize>
J~
C570
0T>
Cntr &
Noat
"""

# Run parser
parsed_data = parse_easyocr_ocr_output(ocr_output)

# Print results
print("\n📋 Formatted OCR Data:\n" + "-"*30)
for key, value in parsed_data.items():
    if isinstance(value, list):
        print(f"{key}:")
        for item in value:
            print(f"  - {item}")
    else:
        print(f"{key}: {value}")
```

```python
def parse_easyocr_misc_output(text):
    lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

    result = {
        "Possible Name/Initials": [],
        "Patient/Case ID": None,
        "Numeric Codes": [],
        "Alphanumeric Codes": [],
        "Center Info": None,
        "Unclassified": []
    }

    for line in lines:
        if line.lower() in ["cntr", "cntr &", "center", "centre", "hospital", "clinic"]:
            result["Center Info"] = line
        elif "cntr" in line.lower() or "cntr &" in line.lower():
            result["Center Info"] = line
        elif line.isdigit() and len(line) >= 5:
            result["Patient/Case ID"] = line
        elif line.isdigit():
            result["Numeric Codes"].append(line)
        elif line.isalnum() and len(line) <= 4:
            result["Possible Name/Initials"].append(line)
        elif any(c.isalnum() for c in line):
            result["Alphanumeric Codes"].append(line)
        else:
            result["Unclassified"].append(line)

    return result


# OCR Input
ocr_output = """
Aec
894271
41-
2
n
4
JV
T~td
647
I_
D"
34
Jtteo
Lize>
J~
C570
0T>
Cntr &
Noat
"""

# Parse and print
parsed = parse_easyocr_misc_output(ocr_output)

print("📋 Parsed OCR Data")
print("-" * 30)
for key, val in parsed.items():
    print(f"{key}:")
    if isinstance(val, list):
        for item in val:
            print(f"  - {item}")
    else:
        print(f"  {val}")
```
